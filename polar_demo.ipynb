
# Polar Demo


Demo of a line plot on a polar axis.



```python
import numpy as np
import matplotlib.pyplot as plt


r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)

ax.plot(1, 1, 'bo')
ax.plot(1, 1.1, 'bo')
ax.plot(1, 1.2, 'bo')
ax.plot(1, 1.3, 'bo')

ax.set_rmax(2)

plt.show()
```
