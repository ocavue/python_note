```python
# reference: https://russelrayphotos2.com/tag/inflorescence/

import numpy as np
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation

from matplotlib.pyplot import figure, show, rc

fig, axes = plt.subplots(1, 1, subplot_kw=dict(polar=True))
ax = axes

# # ax = plt.subplot(111, projection='polar')

# # force square figure and square axes looks better for polar, IMO
# fig = figure(figsize=(8, 8))
# ax = fig.add_axes([100,100,100,100], projection='polar', facecolor='#d5de9c')


def update(m):
    const = 10
    for n in range(1, m):
        radius = const * ( n ** 0.5 )
        theta = n * 7
        # print(radius, theta)
        ax.plot(theta, radius, 'bo')

    plt.show()
    return ax


anim = FuncAnimation(fig, update, frames=np.arange(0, 20), interval=100)
    # plt.show() will just loop the animation forever.
plt.show()
























```
